```python
# Contador de tokens diferentes (Antes de aplicar lematización)

# Extraer directamente los tokens en el texto
Tokens = [tok.text.lower() for tok in Texto if not tok.is_punct | tok.is_stop]

# Eliminar los elementos repetidos de la lista
TokensDif = set(Tokens)

# Extraer las lematizaciones del texto original
Lemmas = [lem.lemma_.lower() for lem in Texto if not lem.is_punct | lem.is_stop]

# Eliminar los elementos repetidos de la lista
LemmasDif = set(Lemmas)

# Impresión de resultados
print("Tokens encontrados en todo el texto: ", len(Texto))
print("Tokens Únicos en todo el texto: ", len(TokensDif))
print("Lemmas Únicos en todo el texto: ", len(LemmasDif))
print("Diferencia entre Tokens y Lemmas:", len(TokensDif) - len(LemmasDif))
```

```python
# Agregamos en un diccionario cada token y su frecuencia de aparición
import matplotlib.pyplot as plt

DicTokens = {}
for Token in TokensDif:
    freq = Tokens.count(Token)
    if freq > 3:
        DicTokens.setdefault(Token, freq)

# Graficamos cada token contra la cantidad de veces que aparece
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(20,5))
plt.bar(DicTokens.keys(), DicTokens.values(), color=['lightblue','steelblue','lightslategrey'])
plt.axhline(sum(DicTokens.values())/len(DicTokens.values()), lw = 3, c='r', xmax=100)
plt.title('Tokens diferentes') 
plt.xticks(rotation=80)
plt.show()
```

```python
# Aregamos en un diccionario cada Lemma y su frecuencia de aparición
DicLemmas = {}
for Lemma in LemmasDif:
    freq = Lemmas.count(Lemma)
    if freq > 3:
        DicLemmas.setdefault(Lemma, freq)
        
# Graficamos cada lemma contra la cantidad de veces que aparece
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(20,5))
plt.bar(DicLemmas.keys(), DicLemmas.values(), color=['lightblue','steelblue','lightslategrey'])
plt.axhline(sum(DicLemmas.values())/len(DicLemmas.values()), lw = 3, c='r', xmax=100)
plt.title('Lemmas diferentes') 
plt.xticks(rotation=80)
plt.show()
```
